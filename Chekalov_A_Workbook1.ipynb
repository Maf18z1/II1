4.3.2 –ó–∞–¥–∞–Ω–∏–µ:
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è sqrt(1 + e^sqrt(x) +cos(x^2))/(abs(1 - sin^3(x))) + ln(abs(2x))
–°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π

```python
import math
import matplotlib.pyplot as plt
def fun(x):
  return math.sqrt(1 + (math.e)**x**0.5 + math.cos(x**2))/(abs(1 - math.sin(x)**3)) + math.log(2*x)
a=[fun(i) for i in range(1,11)]
c=[i for i in range(1,11)]

plt.grid()
plt.plot(c, a)
plt.show()
plt.grid()
plt.scatter(c[:5], a[:5])
```

4.3.5 –ó–∞–¥–∞–Ω–∏–µ:
–°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤–∫–ª—é—á–∞—é—â–∏–π –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è
–¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤ + ‚Ä≤, ‚Ä≤ ‚àí ‚Ä≤, ‚Ä≤ ‚àô ‚Ä≤, ‚Ä≤ √∑ ‚Ä≤, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ
—Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: ùëí^(ùë•+ùë¶)
, sin(ùë• + ùë¶) , cos(ùë• + ùë¶) , ùë•^y

```python
import math
x, y = int(input("x=")), int(input("y="))
n = int(input(' 1. x+y \n 2. x-y \n 3. x*y \n 4. x/y \n 5. e^(x+y) \n 6. sin(x+y) \n 7. cos(x+y) \n 8. x^y \n –û–ø–µ—Ä–∞—Ü–∏—è: '))
match n: # Python 3.10+ required
    case 1:
        print("x + y = ",x+y)
    case 2:
        print("x - y = ",x-y)
    case 3:
        print("x * y = ",x*y)
    case 4:
        try:
            print("x / y = ",x/y)
        except ZeroDivisionError:
            print(x, "—Ä–∞–∑–¥–µ–ª–∏—Ç—å –Ω–∞",y,"–Ω–µ–ª—å–∑—è!")
    case 5:
        print("e^(x + y) = ",math.e**(x+y))
    case 6:
        print("sin(x + y) = ",math.sin(x+y))
    case 7:
        print("cos(x + y) = ",math.cos(x+y))
    case 8:
        print("x^y = ",x**y)
```
